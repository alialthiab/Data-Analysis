```python
import pandas as pd
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv('/content/ford.csv')
```

```python
clt_data=pd.DataFrame()
```

```python
clt_data['pre_data_milage']=data.mileage
```

```python
clt_data['pre_data_milage'][1:11].mean()
```

```python
means_data = np.array([])
```

```python
for i in range(clt_data['pre_data_milage'].shape[0]):
    means_data = np.append(means_data,clt_data['pre_data_milage'][i:i+10].mean())
    i+=10
```

```python
means_data[2]
```

```python
clt_data['post_data_milage'] = pd.Series(means_data)
```

```python
clt_data.hist()
```

```python
sns.distplot(clt_data.post_data_milage)
sns.distplot(clt_data.pre_data_milage)
```

```python
clt_data['pre_random_nums']=pd.DataFrame(np.random.randint(1000, size=(17965)))
```

```python
new_pd_random_nums=np.array([])
for i in range(clt_data.shape[0]):
    new_pd_random_nums = np.append(new_pd_random_nums, clt_data.pre_random_nums[i:i+10].mean())
    i+=10
clt_data['post_random_nums']= pd.DataFrame(new_pd_random_nums)
```
